# Plots

## Import libraries

```python
from __future__ import print_function, division
import numpy as np
import pandas as pd
import matplotlib
matplotlib.rcParams['text.usetex'] = True
import matplotlib.pyplot as plt
```

## Read in training results for BCD and SGD

```python
# BCD training losses
bcd_train_loss = {}
bcd_train_loss_dict = []
for seed in 5,8,10,15,16:
    bcd_train_loss['bcd_train_loss_{0}'.format(seed)] = np.genfromtxt('training_loss_' + str(seed) + '.csv', delimiter=',')
    bcd_train_loss_dict.append('bcd_train_loss_{0}'.format(seed))
bcd_train_loss_matrix = np.array([bcd_train_loss[i] for i in bcd_train_loss_dict])

# SGD training losses
#sgd_train_loss = {}
#sgd_train_loss_dict = []
#for seed in 5,8,10,15,16:
#    sgd_train_loss['sgd_train_loss_{0}'.format(seed)] = np.genfromtxt('sgd_training_loss_' + str(seed) + '.csv', delimiter=',')
#    sgd_train_loss_dict.append('sgd_train_loss_{0}'.format(seed))
#sgd_train_loss_matrix = np.array([sgd_train_loss[i] for i in sgd_train_loss_dict])

# BCD training accuracies
bcd_train_acc = {}
bcd_train_dict = []
for seed in 5,8,10,15,16:
    bcd_train_acc['bcd_acc_train_{0}'.format(seed)] = np.genfromtxt('accuracy_train_' + str(seed) + '.csv', delimiter=',')
    bcd_train_dict.append('bcd_acc_train_{0}'.format(seed))
bcd_train_acc_matrix = np.array([bcd_train_acc[i] for i in bcd_train_dict])
 
# SGD training accuracies
sgd_train_acc = {}
sgd_train_dict = []
for seed in 5,8,10,15,16:
    sgd_train_acc['sgd_acc_train_{0}'.format(seed)] = np.genfromtxt('sgd_accuracy_train_' + str(seed) + '.csv', delimiter=',')
    sgd_train_dict.append('sgd_acc_train_{0}'.format(seed))
sgd_train_acc_matrix = np.array([sgd_train_acc[i] for i in sgd_train_dict])
```

## Read in testing results for BCD and SGD

```python
# BCD testing accuracies
bcd_test_acc = {}
bcd_test_dict = []
for seed in 5,8,10,15,16:
    bcd_test_acc['bcd_acc_test_{0}'.format(seed)] = np.genfromtxt('accuracy_test_' + str(seed) + '.csv', delimiter=',')
    bcd_test_dict.append('bcd_acc_test_{0}'.format(seed))
bcd_test_acc_matrix = np.array([bcd_test_acc[i] for i in bcd_test_dict])

# SGD testing accuracies
sgd_test_acc = {}
sgd_test_dict = []
for seed in 5,8,10,15,16:
    sgd_test_acc['sgd_acc_test_{0}'.format(seed)] = np.genfromtxt('sgd_accuracy_test_' + str(seed) + '.csv', delimiter=',')
    sgd_test_dict.append('sgd_acc_test_{0}'.format(seed))
sgd_test_acc_matrix = np.array([sgd_test_acc[i] for i in sgd_test_dict])
```

## Compute the mean accuracies of 5 runs

```python
bcd_train_loss_mean = np.mean(bcd_train_loss_matrix, axis=0)
#sgd_train_loss_mean = np.mean(sgd_train_loss_matrix, axis=0)

bcd_train_acc_mean = np.mean(bcd_train_acc_matrix, axis=0)
sgd_train_acc_mean = np.mean(sgd_train_acc_matrix, axis=0)

bcd_test_acc_mean = np.mean(bcd_test_acc_matrix, axis=0)
sgd_test_acc_mean = np.mean(sgd_test_acc_matrix, axis=0)
```

## Plots of training losses

### BCD

```python
# Plot of BCD Training Loss 
fig_train_loss_bcd = plt.figure(1)
plt.plot(np.arange(1, len(bcd_train_loss_mean[0:49]) + 1), bcd_train_loss_mean[0:49], label='BCD', linewidth=1.5)
plt.title('Ten-layer MLP (BCD)')
plt.xlabel('Epochs')
plt.ylabel('Training loss')
plt.legend()
```

## Plots of training accuracies

```python
# Plot of Accuracies
fig_train_accuracy = plt.figure(3)
plt.plot(np.arange(1, len(bcd_train_acc_mean[0:99]) + 1), bcd_train_acc_mean[0:99], label='BCD', linewidth=1.5)
plt.plot(np.arange(1, len(sgd_train_acc_mean[0:99]) + 1), sgd_train_acc_mean[0:99], label='SGD', linewidth=1.5)
plt.title('Ten-layer MLP')
plt.xlabel('Epochs')
plt.ylabel('Training accuracy')
plt.legend()
```

## Plots of testing accuracies

```python
# Plot of Accuracies
fig_test_accuracy = plt.figure(4)
plt.plot(np.arange(1, len(bcd_test_acc_mean[0:99]) + 1), bcd_test_acc_mean[0:99], label='BCD', linewidth=1.5)
plt.plot(np.arange(1, len(sgd_test_acc_mean[0:99]) + 1), sgd_test_acc_mean[0:99], label='SGD', linewidth=1.5)
plt.title('Ten-layer MLP')
plt.xlabel('Epochs')
plt.ylabel('Testing accuracy')
plt.legend()
```

## Plots of BCD

```python
# Plot of BCD Accuracies
fig_bcd_accuracy = plt.figure(5)
plt.plot(np.arange(1, len(bcd_train_acc_mean[0:99]) + 1), bcd_train_acc_mean[0:99], label='Train', linewidth=1.5)
plt.plot(np.arange(1, len(bcd_test_acc_mean[0:99]) + 1), bcd_test_acc_mean[0:99], label='Test', linewidth=1.5)
plt.title('BCD Ten-layer MLP')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
```
